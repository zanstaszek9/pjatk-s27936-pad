4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
import pandas as pd
import numpy as np

ex1_data_csv = pd.read_csv('orders.csv')
ex1_data_csv
# ex1_data_csv.isna().sum()
```

```python
### 1 Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;

# ex1_data_csv['order_date'] ## print jako 2016/09/04

ex1_data_csv['order_date'] = pd.to_datetime(ex1_data_csv['order_date'])

ex1_data_csv['order_date'] ## print jako 2016-09-04
```

```python
### 2 Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
ex1_data_csv['tshirt_category'].nunique()
ex1_data_csv['tshirt_category'].unique()

print(str(ex1_data_csv['tshirt_category'].nunique()) + ' ||| ' + str(ex1_data_csv['tshirt_category'].unique()))
```

```python
### 3 Wyczyść kolumnę tshirt_category normalizując nazwy produktów. 
###   Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, 
###   np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

ex1_data_csv['tshirt_category'].unique()

ex1_data_csv['tshirt_category'] = ex1_data_csv['tshirt_category'].apply(
    lambda x: x
    .replace('Tshirt ', 'T-Shirt ')
    .replace('Wh ', 'White ')
    .replace('Bl ', 'Black ')
  )

ex1_data_csv['tshirt_category'].unique()

```

```python
### 4 Rozdziel kolumnę tshirt category na 3 kolumny: 
###   tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, 
###   tshirt_size, która będzie informowała o rozmiarze koszulki, 
###   i tshirt_colour, która będzie informowała o kolorze. 
###   Skorzystaj z metody apply.


# dodanie dodatkowych danych do wartości, których nie da się rozdzielic na 3 (Hoodie, Tennis Shirt)
ex1_data_csv['tshirt_category'] = ex1_data_csv['tshirt_category'].apply(
    lambda x: x
    .replace('Hoodie', 'Default Hoodie U') # 'U' od Unisex
    .replace('Tennis Shirt', 'Default Tennis Shirt U')  # 'U' od Unisex
  )




ex1_data_csv['tshirt_type'] = ex1_data_csv['tshirt_category'].apply(
    lambda x: ' ' .join(x
    .split(' ')[1:-1] # 'Tennis Shirt is a Type name made of two elements, therefore [1:-1]
    )
  )


ex1_data_csv['tshirt_size'] = ex1_data_csv['tshirt_category'].apply(
    lambda x: x
    .split(' ')[-1]
  )


ex1_data_csv['tshirt_colour'] = ex1_data_csv['tshirt_category'].apply(
    lambda x: x
    .split(' ')[0]
  )

ex1_data_csv['tshirt_category'].unique()





```
